```python



word_map = {            
    1: "BOOK", 2: "CHILD", 3: "GOLD", 4: "HOTEL", 5: "KING", 6: "MARKET", 7: "PAPER", 8: "RIVER", 9: "SKIN", 10: "TREE",
    11: "BUTTER", 12: "COLLEGE", 13: "DOLLAR", 14: "EARTH", 15: "FLAG", 16: "HOME", 17: "MACHINE", 18: "OCEAN", 19: "SKY", 20: "WIFE",
    21: "BLOOD", 22: "CORNER", 23: "ENGINE", 24: "GIRL", 25: "HOUSE", 26: "LETTER", 27: "ROCK", 28: "SHOES", 29: "VALLEY", 30: "WOMAN",
    31: "BABY", 32: "CHURCH", 33: "DOCTOR", 34: "FIRE", 35: "GARDEN", 36: "PALACE", 37: "SEA", 38: "TABLE", 39: "VILLAGE", 40: "WATER"
}

# Define the word lists
list1 = ["BOOK", "CHILD", "GOLD", "HOTEL", "KING", "MARKET", "PAPER", "RIVER", "SKIN", "TREE"]
list11 = ["BUTTER", "COLLEGE", "DOLLAR", "EARTH", "FLAG", "HOME", "MACHINE", "OCEAN", "SKY", "WIFE"]
list21 = ["BLOOD", "CORNER", "ENGINE", "GIRL", "HOUSE", "LETTER", "ROCK", "SHOES", "VALLEY", "WOMAN"]
list31 = ["BABY", "CHURCH", "DOCTOR", "FIRE", "GARDEN", "PALACE", "SEA", "TABLE", "VILLAGE", "WATER"]

def process_column(series):
    result = []
    for num in series:
        num_str = str(num).strip()

        if not num_str:
            result.append(0)
            continue

        try:
            number = int(num_str)
        except ValueError:
            result.append(0)
            continue

        if number not in range(1, 41):
            result.append(0)
            continue

        word = word_map.get(number)

        if number in range(1, 11):
            if word in list1:
                index = list1.index(word) + 1
                result.append(index)
            else:
                result.append(-1)
        elif number in range(11, 21):
            if word in list11:
                index = list11.index(word) + 1
                result.append(index)
            else:
                result.append(-1)
        elif number in range(21, 31):
            if word in list21:
                index = list21.index(word) + 1
                result.append(index)
            else:
                result.append(-1)
        elif number in range(31, 41):
            if word in list31:
                index = list31.index(word) + 1
                result.append(index)
            else:
                result.append(-1)
    return result



# List of columns to process
columns_to_process = [f"PD182M{i}" for i in range(1, 19)]

processed_columns = {}

# Loop through each column, apply the function, and store the result
for col in columns_to_process:
    processed_data = process_column(new_df[col])
    processed_columns[col] = processed_data

new_result_df = pd.DataFrame(processed_columns)

new_result_df.head(10)
```
