```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/mnt/home/tranha1/research/cbcclab/git_repos/experiments/Tran25longitudinal.git/CognitionData/demsHRS.csv')
df
```

```python
#Create dataframe containing only the columns we are dealing with
subset = df[["age96","age98","age00","age02","age04","age06","age08","age10","age12","age14","age16","age18","age20","age22"]]
```

```python
# Replace this with your actual DataFrame if not already
df = subset.copy()  # Copy and create a separate data frame

# Convert from strings to numeric, any error to NA values
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
# Create 14 empty lists (1 list per visit)
visits = []
 
for i in range(14):
    listinit = []
    visits.append(listinit)
    
for _, row in df.iterrows(): #for each row in the dataframe, iterate
    #set visit counter to 0 for each person
    visit_number = 0  
    # row is a list of age of that person for each visit
    for age in row:
        if pd.notna(age):  # Check it's a real age (empty/NA)
            # check which visit to add the age to its corresponding visit list
            if visit_number < 14:
                # append the age of that visit to the visit number
                visits[visit_number].append(age)
                # increment the counter by 1 for the next encounter of the visit
                visit_number += 1
```

```python
for i in range(14):
    print(f"Visit {i+1}: {len(visits[i])} values, example: {visits[i][:5]}")
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(18, 20))
axes = axes.flatten()

for i in range(14):
    ax = axes[i]
    if visits[i]:  # Only plot if there's data
        ax.hist(visits[i], bins=range(18, 109, 2), color='skyblue', edgecolor='black')
    ax.set_title(f'Visit {i+1}')
    ax.set_xlabel('Age')
    ax.set_ylabel('Count')

# Hide the last subplot if extra
if len(axes) > 14:
    axes[14].axis('off')

plt.tight_layout()
plt.suptitle('Age Distribution by Visit Number', fontsize=20, y=1.02)
plt.show()

```
