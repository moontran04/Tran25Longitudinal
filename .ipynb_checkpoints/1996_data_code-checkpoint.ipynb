```python
import pandas as pd

# #Define the 4 word lists
# list1 = [  
#     "HOTEL", "RIVER", "TREE", "SKIN", "GOLD",
#     "MARKET", "PAPER", "CHILD", "KING", "BOOK"
# ]

# list11 = [
#     "SKY", "OCEAN", "FLAG", "DOLLAR", "WIFE",
#     "MACHINE", "HOME", "EARTH", "COLLEGE", "BUTTER"
# ]

# list21 = [
#     "WOMAN", "ROCK", "BLOOD", "CORNER", "SHOES",
#     "LETTER", "GIRL", "HOUSE", "VALLEY", "ENGINE"
# ]

# list31 = [
#     "WATER", "CHURCH", "DOCTOR", "PALACE", "FIRE",
#     "GARDEN", "SEA", "VILLAGE", "BABY", "TABLE"
# ]

def process_column(series):
    result = []
    for num in series:
        num_str = str(num).strip()

        if not num_str:
            result.append(0)
            continue

        try:
            number = int(num_str)
        except ValueError:
            result.append(0)
            continue

        if number not in range(1, 41):
            result.append(0)
            continue

        word = word_map.get(number)

        if number in range(1, 11):
            if word in list1:
                index = list1.index(word) + 1
                result.append(index)
            else:
                result.append(-1)
        elif number in range(11, 21):
            if word in list11:
                index = list11.index(word) + 1
                result.append(index)
            else:
                result.append(-1)
        elif number in range(21, 31):
            if word in list21:
                index = list21.index(word) + 1
                result.append(index)
            else:
                result.append(-1)
        elif number in range(31, 41):
            if word in list31:
                index = list31.index(word) + 1
                result.append(index)
            else:
                result.append(-1)
    return result



# List of columns to process
columns_to_process = [f"E1174M{i}" for i in range(1, 12)]

processed_columns = {}

# Loop through each column, apply the function, and store the result
for col in columns_to_process:
    processed_data = process_column(new_df[col])
    processed_columns[col] = processed_data

new_result_df = pd.DataFrame(processed_columns)

new_result_df.head(10)
```

```python
import pandas as pd
# word_map = {            
#     1: "BOOK",
#     2: "CHILD",
#     3: "GOLD",
#     4: "HOTEL",
#     5: "KING",
#     6: "MARKET",
#     7: "PAPER",
#     8: "RIVER",
#     9: "SKIN",
#     10: "TREE",
#     11: "BUTTER",
#     12: "COLLEGE",
#     13: "DOLLAR",
#     14: "EARTH",
#     15: "FLAG",
#     16: "HOME",
#     17: "MACHINE",
#     18: "OCEAN",
#     19: "SKY",
#     20: "WIFE",
#     21: "BLOOD",
#     22: "CORNER",
#     23: "ENGINE",
#     24: "GIRL",
#     25: "HOUSE",
#     26: "LETTER",
#     27: "ROCK",
#     28: "SHOES",
#     29: "VALLEY",
#     30: "WOMAN",
#     31: "BABY",
#     32: "CHURCH",
#     33: "DOCTOR",
#     34: "FIRE",
#     35: "GARDEN",
#     36: "PALACE",
#     37: "SEA",
#     38: "TABLE",
#     39: "VILLAGE",
#     40: "WATER"
# }

# #Define the 4 word lists
# list1 = [  
#     "HOTEL", "RIVER", "TREE", "SKIN", "GOLD",
#     "MARKET", "PAPER", "CHILD", "KING", "BOOK"
# ]

# list11 = [
#     "SKY", "OCEAN", "FLAG", "DOLLAR", "WIFE",
#     "MACHINE", "HOME", "EARTH", "COLLEGE", "BUTTER"
# ]

# list21 = [
#     "WOMAN", "ROCK", "BLOOD", "CORNER", "SHOES",
#     "LETTER", "GIRL", "HOUSE", "VALLEY", "ENGINE"
# ]

# list31 = [
#     "WATER", "CHURCH", "DOCTOR", "PALACE", "FIRE",
#     "GARDEN", "SEA", "VILLAGE", "BABY", "TABLE"
# ]

def calculate_participant_probability(visit_file_path, participant_id, trial_columns, id_column="ID"):
    """
    Given a visit file, participant ID, and trial columns, return the probability of correct guesses.

    Args:
        visit_file_path (str): Path to the visit CSV file.
        participant_id (int or str): The ID of the participant.
        trial_columns (list of str): List of column names containing the trials.
        id_column (str): Name of the column that contains participant IDs (default is "ID").

    Returns:
        float or None: Probability value (0.0 to 1.0), or None if something went wrong.
    """
    try:
        df = pd.read_csv(visit_file_path)
    except FileNotFoundError:
        print(f"File not found: {visit_file_path}")
        return None

    # Make sure the ID column exists
    if id_column not in df.columns:
        print(f"ID column '{id_column}' not found in file.")
        return None

    # Make sure the participant exists
    if participant_id not in df[id_column].values:
        print(f"Participant ID {participant_id} not found in file.")
        return None

    # Select the participant's row
    participant_row = df[df[id_column] == participant_id]

    # Check trial columns exist
    if not all(col in participant_row.columns for col in trial_columns):
        print("One or more trial columns are missing from the file.")
        return None

    trials = participant_row[trial_columns].iloc[0]

    # Process trial responses into positions
    processed = []
    for num in trials:
        num_str = str(num).strip()
        if not num_str:
            processed.append(0)
            continue
        try:
            number = int(num_str)
        except ValueError:
            processed.append(0)
            continue
        if number not in word_map:
            processed.append(0)
            continue
        word = word_map[number]
        if number in range(1, 11):
            processed.append(list1.index(word) + 1 if word in list1 else -1)
        elif number in range(11, 21):
            processed.append(list11.index(word) + 1 if word in list11 else -1)
        elif number in range(21, 31):
            processed.append(list21.index(word) + 1 if word in list21 else -1)
        elif number in range(31, 41):
            processed.append(list31.index(word) + 1 if word in list31 else -1)

    # Calculate probability: unique valid values from first 10 trials
    first_10 = processed[:10]
    unique_valid = set(val for val in first_10 if val > 0)
    probability = len(unique_valid) / 10

    return probability
```
